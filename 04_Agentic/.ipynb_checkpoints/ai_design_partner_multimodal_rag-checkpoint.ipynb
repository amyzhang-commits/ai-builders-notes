This project emerged from the failure to connect local offline models to online MCP servers. This became an opportunity to rethink tools beyond remote API calls. The question became: **"What if instead of giving the LLM a search tool, you gave them a pen?"**

## The Tool-Calling Breakthrough

Instead of traditional API calls, we created "drawing tools" that the AI can invoke:
```
json
{
  "text_reasoning": "Based on cooling strategies with 3 retrieved chunks including 8 technical diagrams...",
  "actions": [
    {"mode": "structured", "shape": "rectangle", "coords": [10,10,25,80], "reasoning": "Primary facade mass - based on architectural diagrams"},
    {"mode": "scribble", "points": [[11,75],[15,70],[20,65]], "style": "wavy", "reasoning": "Cooling fins/shading - inspired by technical drawings"}
  ]
}
```
Each "action" is essentially a tool call to drawing functions:

- draw_rectangle(coordinates, reasoning)
- draw_scribble(points, style, reasoning)

## Next Steps
- Ultimately, the goal is to have collaborative design sessions with an AI design partner who can respond to your ideas with their own. Rather than returning only text-based chains of reasoning, the AI shares the equivalent in diagrammatic form - building up visual ideas through conversation, hopefully with increasing free-form drawing capabilities.
- Result: A system where you can say "I'm thinking the fa√ßade should actively cool the tower" and get both thoughtful analysis AND visual sketches that build on each other, informed by actual architectural knowledge and technical diagrams.

### üìù Textbook has 1.3 million characters.

````python
# 2) Load and Extract PDF Text
import PyPDF2
# or try: import pdfplumber  # alternative if PyPDF2 doesn't work well

pdf_path = Path("/Users/amyzhang/Documents/Projects/saluda-rag/pdfs/Constructing_Architecture.pdf")

# Function to extract text from PDF
def extract_pdf_text(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text() + "\n"
    return text

# Extract the text
book_text = extract_pdf_text(pdf_path)
print(f"Extracted {len(book_text)} characters from PDF")
print("First 500 characters:", book_text[:500])
````

## 3. Chunk the text for RAG.

````python
# 3) Text Chunking for RAG
def chunk_text(text, chunk_size=800, overlap=100):
    """Break text into overlapping chunks for better retrieval"""
    words = text.split()
    chunks = []
    i = 0
    while i < len(words):
        chunk = " ".join(words[i:i+chunk_size])
        chunks.append(chunk)
        i += chunk_size - overlap  # Move forward, but overlap some words
    return chunks

# Create chunks from your architecture book
chunks = chunk_text(book_text, chunk_size=800, overlap=100)

print(f"Created {len(chunks)} chunks from the book")
print(f"Average chunk length: {sum(len(c) for c in chunks) / len(chunks):.0f} characters")
print("\nFirst chunk preview:")
print(chunks[0][:300] + "...")
print("\nSecond chunk preview:")  
print(chunks[1][:300] + "...")
````

---
### üìù Understanding 'overlap = 100'
> - The end of chunk 100 and the beginning of chunk 101 have the same text:

"...so far had mainly been used as ornamentation, could establish itself as a veritable building material. The huge church windows also showed glass to be a com-plementary building material..."

> - This ensures that a query re: glass as a building material will be able to bring both chunks to the RAG model's attention and consideration.
> - Here, had the chunk left this out in each, we would lose the full trajectory of glass from ornament to structural material. **Conceptual Continuity** is preserved.

## 4. TF-IDF ('Term Frequency - Inverse Document Frequency) Search Index
> Beyond simple word finding: TF-IDF turns each piece of text into a vector of numbers, with common words like "the" and "of" having lower value; represented in this way, each chunk can then be mapped as a vector. Cosine similarity then finds which are closest to each other to return a more semantically-informed response.

````python
# 4) Build TF-IDF Search Index
print("Building search index...")
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix = vectorizer.fit_transform(chunks)
print(f"Built index with {tfidf_matrix.shape[0]} documents and {tfidf_matrix.shape[1]} features")
print("Ready for semantic search! üöÄ")
````

## 5. Test RAG Search

````python
# 5) Test the RAG Search
def retrieve_relevant_chunks(query, vectorizer, tfidf_matrix, chunks, k=3):
    """Find the most relevant chunks for a given query"""
    # Convert query to same vector space
    query_vector = vectorizer.transform([query])
    
    # Calculate similarity with all chunks
    similarities = cosine_similarity(query_vector, tfidf_matrix)[0]
    
    # Get top k most similar chunks
    top_indices = similarities.argsort()[::-1][:k]
    
    results = []
    for i, idx in enumerate(top_indices):
        results.append({
            'chunk_id': idx,
            'similarity_score': similarities[idx],
            'content': chunks[idx][:500] + "..." if len(chunks[idx]) > 500 else chunks[idx]
        })
    
    return results

# Test it with an architecture query!
test_query = "cooling strategies for building facades"
results = retrieve_relevant_chunks(test_query, vectorizer, tfidf_matrix, chunks, k=3)

print(f"Query: '{test_query}'")
print("="*50)
for i, result in enumerate(results, 1):
    print(f"\n{i}. Chunk {result['chunk_id']} (Score: {result['similarity_score']:.3f})")
    print(result['content'])
    print("-" * 30)
````

# 2Ô∏è‚É£ Part II. Multi-Modal RAG 
> - We noticed that chunk 3 in our first test referred to a diagram. This led us to ask whether we could build a RAG search that could distinguish between text pertaining to passage content and text pertaining to diagrams.
> - In future conversation, the LLM would be able to make the distinction and draw upon these resources differently.
> - Moreover, the diagrams might even help enrich the model's sketching process (the tools that are called in relation to the chain-of-reasoning).
> - Gemma3n is capable of multi-modal analysis. (See 'https://gemma-llm.readthedocs.io/en/latest/colab_multimodal.html'). **Gemma3n is not yet integrated - we're using a mock model to test the multi-modal retrieval flow.**
> - **Test queries:** 
        - "cooling strategies for building facades"
        - "structural beam connections" 
        - "glass facade systems with diagrams"
> - For this mock model, the generated visualization is **a rule-based translation** of the model's chain of reasoning. These can be considered a "tool" that a function can call. 
> - **Result:** The system now retrieves chunks with associated diagrams (e.g., "including 21 technical diagram(s)") and generates contextually appropriate drawing actions like "cooling fins/shading - inspired by technical drawings."

````python
# Multi-Modal RAG Architecture Design Tool
# Now with vision capabilities! üöÄ

import json
import os
import math
import random
from pathlib import Path
import fitz  # PyMuPDF for better PDF handling
from PIL import Image
import io
import base64

# For multi-modal embeddings - we'll use a simple approach first
# You might want to upgrade to more sophisticated vision models later

# -------------------------
# 1. MULTI-MODAL PDF EXTRACTION
# -------------------------

def extract_pdf_content_with_images(pdf_path, output_dir="pdf_images"):
    """Extract both text and images from PDF, keeping them associated"""
    
    # Create output directory for images
    output_dir = Path(output_dir)
    output_dir.mkdir(exist_ok=True)
    
    # Open PDF
    doc = fitz.open(pdf_path)
    content_chunks = []
    
    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        
        # Extract text from page
        text = page.get_text()
        
        # Extract images from page
        image_list = page.get_images()
        page_images = []
        
        for img_index, img in enumerate(image_list):
            # Get image data
            xref = img[0]
            base_image = doc.extract_image(xref)
            image_bytes = base_image["image"]
            image_ext = base_image["ext"]
            
            # Save image
            image_filename = f"page_{page_num+1}_img_{img_index+1}.{image_ext}"
            image_path = output_dir / image_filename
            
            with open(image_path, "wb") as img_file:
                img_file.write(image_bytes)
            
            page_images.append(str(image_path))
        
        # Create content chunk with text and associated images
        if text.strip() or page_images:  # Only add if there's content
            content_chunks.append({
                "page": page_num + 1,
                "text": text.strip(),
                "images": page_images,
                "type": "page_content"
            })
    
    doc.close()
    return content_chunks

# -------------------------
# 2. MULTI-MODAL CHUNKING
# -------------------------

def create_multimodal_chunks(content_chunks, chunk_size=800, overlap=100):
    """Create chunks that preserve text-image relationships"""
    
    chunks = []
    chunk_id = 0
    
    for content in content_chunks:
        text = content["text"]
        images = content["images"]
        page = content["page"]
        
        if not text.strip():
            # If page has only images, create image-only chunk
            if images:
                chunks.append({
                    "id": chunk_id,
                    "page": page,
                    "text": f"[Page {page} contains {len(images)} diagram(s)/figure(s)]",
                    "images": images,
                    "type": "image_heavy"
                })
                chunk_id += 1
            continue
        
        # Split text into overlapping chunks like before
        words = text.split()
        i = 0
        while i < len(words):
            chunk_text = " ".join(words[i:i+chunk_size])
            
            # Associate images with text chunks from the same page
            chunk_images = images if images else []
            
            chunks.append({
                "id": chunk_id,
                "page": page,
                "text": chunk_text,
                "images": chunk_images,
                "type": "text_with_images" if chunk_images else "text_only"
            })
            
            chunk_id += 1
            i += chunk_size - overlap
    
    return chunks

# -------------------------
# 3. SIMPLE MULTI-MODAL RETRIEVAL
# -------------------------

def retrieve_multimodal_chunks(query, chunks, k=3):
    """
    Simple multi-modal retrieval - we'll score based on:
    1. Text similarity (like before)
    2. Presence of images (boost for visual content)
    3. Keywords that suggest visual content
    """
    
    # Keywords that suggest the user wants visual/technical content
    visual_keywords = ["diagram", "figure", "drawing", "section", "detail", "construction", 
                      "structure", "facade", "system", "connection", "assembly"]
    
    scored_chunks = []
    query_lower = query.lower()
    
    for chunk in chunks:
        score = 0
        text = chunk["text"].lower()
        
        # Basic text similarity (simplified - count matching words)
        query_words = set(query_lower.split())
        text_words = set(text.split())
        word_overlap = len(query_words.intersection(text_words))
        score += word_overlap * 10
        
        # Boost for having images
        if chunk["images"]:
            score += 20
            
        # Extra boost if query suggests visual content and chunk has images
        if any(keyword in query_lower for keyword in visual_keywords) and chunk["images"]:
            score += 30
            
        # Boost for relevant content
        if any(word in text for word in query_words):
            score += 5
            
        scored_chunks.append((score, chunk))
    
    # Sort by score and return top k
    scored_chunks.sort(key=lambda x: x[0], reverse=True)
    return [chunk for _, chunk in scored_chunks[:k]]

# -------------------------
# 4. CONVERSATION & CANVAS (same as before)
# -------------------------

class Conversation:
    def __init__(self):
        self.messages = []
    def append(self, role, text):
        self.messages.append({"role": role, "text": text})
    def last_user(self):
        for m in reversed(self.messages):
            if m["role"] == "user":
                return m["text"]
        return ""

class Canvas:
    def __init__(self):
        self.history = []
    def add_structured(self, shape, coords, reasoning):
        entry = {"mode": "structured", "shape": shape, "coords": coords, "reasoning": reasoning}
        self.history.append(entry)
    def add_scribble(self, points, style, reasoning):
        entry = {"mode": "scribble", "points": points, "style": style, "reasoning": reasoning}
        self.history.append(entry)
    def summary(self, last_n=6):
        out = []
        for s in self.history[-last_n:]:
            if s["mode"] == "structured":
                out.append(f"{s['shape']} at {s['coords']} ({s['reasoning']})")
            else:
                out.append(f"scribble {s['style']} pts:{len(s['points'])} ({s['reasoning']})")
        return "\n".join(out)
    def replay(self):
        print("\n--- Canvas Replay ---")
        for i, s in enumerate(self.history, 1):
            print(f"{i}. {s}")
        print("--- end ---\n")

# -------------------------
# 5. ENHANCED MOCK MODEL (with image awareness)
# -------------------------

def multimodal_mock_model(user_input, retrieved_chunks, canvas_summary):
    """
    Enhanced mock model that's aware of diagrams and figures
    """
    
    # Analyze retrieved content
    has_images = any(chunk["images"] for chunk in retrieved_chunks)
    image_count = sum(len(chunk["images"]) for chunk in retrieved_chunks)
    
    # Create reasoning that incorporates visual content
    text_content = " | ".join([chunk["text"][:100] for chunk in retrieved_chunks])
    
    reasoning = f"Based on '{user_input}' with {len(retrieved_chunks)} retrieved chunks"
    if has_images:
        reasoning += f" including {image_count} technical diagram(s)"
    reasoning += f". Key content: {text_content[:200]}..."
    
    # Generate actions informed by multimodal content
    actions = []
    
    # If we have cooling/facade content, create relevant shapes
    if "cool" in user_input.lower() or "facade" in user_input.lower():
        actions.append({
            "mode": "structured", 
            "shape": "rectangle", 
            "coords": [10, 10, 25, 80], 
            "reasoning": "Primary facade mass - based on architectural diagrams"
        })
        
        if has_images:
            actions.append({
                "mode": "scribble", 
                "points": [[11, 75], [15, 70], [20, 65], [24, 60]], 
                "style": "wavy", 
                "reasoning": "Cooling fins/shading - inspired by technical drawings"
            })
    
    # If structural content, show structural elements
    elif "structure" in user_input.lower() or "beam" in user_input.lower():
        actions.append({
            "mode": "structured", 
            "shape": "rectangle", 
            "coords": [30, 40, 70, 45], 
            "reasoning": "Primary beam - from structural diagrams"
        })
        actions.append({
            "mode": "structured", 
            "shape": "rectangle", 
            "coords": [35, 35, 40, 50], 
            "reasoning": "Column connection - detail from figures"
        })
    
    else:
        # Default architectural response
        actions.append({
            "mode": "structured", 
            "shape": "rectangle", 
            "coords": [20, 20, 40, 70], 
            "reasoning": "Base architectural element"
        })
    
    return json.dumps({
        "text_reasoning": reasoning,
        "actions": actions
    })

# -------------------------
# 6. INTEGRATION & TEST
# -------------------------

def multimodal_design_round(user_text, chunks, conv, canvas):
    """Run a complete multimodal design conversation turn"""
    
    print(f"\nüéØ User: {user_text}")
    conv.append("user", user_text)
    
    # Retrieve relevant chunks (text + images)
    print("üîç Retrieving relevant content...")
    retrieved = retrieve_multimodal_chunks(user_text, chunks, k=3)
    
    # Show what we found
    print(f"üìö Found {len(retrieved)} relevant chunks:")
    for i, chunk in enumerate(retrieved, 1):
        image_info = f" + {len(chunk['images'])} image(s)" if chunk['images'] else ""
        print(f"   {i}. Page {chunk['page']}: {chunk['text'][:100]}...{image_info}")
    
    # Generate AI response
    print("ü§ñ Generating design response...")
    canvas_summary = canvas.summary(last_n=4)
    raw_response = multimodal_mock_model(user_text, retrieved, canvas_summary)
    
    try:
        response = json.loads(raw_response)
        
        # Add reasoning to conversation
        reasoning = response.get("text_reasoning", "")
        conv.append("ai", reasoning)
        print(f"üí≠ AI Reasoning: {reasoning}")
        
        # Execute drawing actions
        print("‚úèÔ∏è Executing drawing actions:")
        for action in response.get("actions", []):
            if action.get("mode") == "structured":
                canvas.add_structured(action.get("shape"), action.get("coords"), action.get("reasoning"))
                print(f"   üìê {action.get('shape')} at {action.get('coords')}: {action.get('reasoning')}")
            else:
                canvas.add_scribble(action.get("points"), action.get("style"), action.get("reasoning"))
                print(f"   ‚úèÔ∏è {action.get('style')} scribble: {action.get('reasoning')}")
        
    except Exception as e:
        print(f"‚ùå Error parsing AI response: {e}")

# -------------------------
# 7. DEMO SETUP
# -------------------------

def setup_multimodal_demo(pdf_path):
    """Set up the complete multimodal system"""
    
    print("üöÄ Setting up Multi-Modal Architecture RAG System...")
    
    # Step 1: Extract content with images
    print("üìñ Extracting PDF content and images...")
    content_chunks = extract_pdf_content_with_images(pdf_path)
    print(f"   Extracted content from {len(content_chunks)} pages")
    
    # Step 2: Create multi-modal chunks
    print("üß© Creating multi-modal chunks...")
    chunks = create_multimodal_chunks(content_chunks, chunk_size=800, overlap=100)
    
    text_chunks = [c for c in chunks if c["type"] in ["text_only", "text_with_images"]]
    image_chunks = [c for c in chunks if c["images"]]
    
    print(f"   Created {len(chunks)} total chunks:")
    print(f"   - {len(text_chunks)} text chunks")
    print(f"   - {len(image_chunks)} chunks with images")
    
    # Step 3: Initialize conversation system
    print("üí¨ Initializing conversation system...")
    conv = Conversation()
    canvas = Canvas()
    
    print("‚úÖ Multi-modal system ready!")
    return chunks, conv, canvas

# -------------------------
# 8. READY TO USE!
# -------------------------

if __name__ == "__main__":
    # Test with your PDF
    pdf_path = "/Users/amyzhang/Documents/Projects/saluda-rag/pdfs/Constructing_Architecture.pdf"
    chunks, conv, canvas = setup_multimodal_demo(pdf_path)
    
    # Try some test queries!
    test_queries = [
        "cooling strategies for building facades",
        "structural beam connections", 
        "glass facade systems with diagrams"
    ]
    
    for query in test_queries:
        multimodal_design_round(query, chunks, conv, canvas)
        print("\n" + "="*60 + "\n")
    
    # Show final canvas
    canvas.replay()
````

````python
# üé® CANVAS VISUALIZER + SAVE
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def visualize_canvas(canvas, title="Architecture Design Canvas"):
    """Simple visualization of the canvas state"""
    
    fig, ax = plt.subplots(1, 1, figsize=(12, 8))
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)
    ax.set_aspect('equal')
    ax.invert_yaxis()  # Flip Y axis for drawing coordinates
    ax.set_title(title, fontsize=16, fontweight='bold')
    
    colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray']
    
    for i, action in enumerate(canvas.history):
        color = colors[i % len(colors)]
        
        if action['mode'] == 'structured':
            shape = action['shape']
            coords = action['coords']
            
            if shape == 'rectangle':
                x1, y1, x2, y2 = coords
                width = x2 - x1
                height = y2 - y1
                rect = patches.Rectangle((x1, y1), width, height, 
                                       linewidth=2, edgecolor=color, 
                                       facecolor=color, alpha=0.3)
                ax.add_patch(rect)
            
            # Add label
            ax.text(coords[0], coords[1] - 3, f"{i+1}: {action['reasoning'][:30]}...", 
                   fontsize=8, color=color, fontweight='bold')
        
        elif action['mode'] == 'scribble':
            points = action['points']
            if points:
                xs = [p[0] for p in points]
                ys = [p[1] for p in points]
                ax.plot(xs, ys, color=color, linewidth=3, alpha=0.7, marker='o')
                
                # Add label
                ax.text(xs[0], ys[0] - 3, f"{i+1}: {action['reasoning'][:30]}...", 
                       fontsize=8, color=color, fontweight='bold')
    
    ax.grid(True, alpha=0.3)
    ax.set_xlabel("X Coordinate")
    ax.set_ylabel("Y Coordinate")
    
    plt.tight_layout()
    return fig  # Return the figure object!

# Create the visualization AND save it
fig = visualize_canvas(canvas, "Multi-Modal Architecture Design Session")

# Save the plot
save_path = f'/Users/amyzhang/Documents/Projects/ai-builders-notes/04_Agentic/assets/sketch1.png'
fig.savefig(save_path, dpi=300, bbox_inches='tight')

# Show it
plt.show()

# Confirmation
print(f'‚úÖ Saved: {save_path}')
````
